```python
import pandas as pd
import numpy as np
```

Abaixo estou importando o .csv que vamos trabalhar e como o cabeçalho esta na linha 0 eu coloco o header como 0 assim não preciso definir as colunas:

```python
df=pd.read_csv('./data/Video_Games.csv',header=0)
df.head(10)
```

```python
display(df)
```

O arquivo .csv trás as vendas de jogos de video game desde os anos 80 até o anos 2010, boscamos analizar os totais de vendas, mais as notas de critica e as notas dos usuários, podendo analizar quais as plataformas e generos mais pupulares.

```python
df.info()
```

Abaixo vamos eliminar as colunas 'NA_Sales','EU_Sales','JP_Sales','Other_Sales','Critic_Count','User_Count', 'Rating' já que vendas separadas não é interessante nesse momento e nem os numeros de criticos e usuários bem como a classificação dos jogos. 

```python
df = df.drop(['NA_Sales','EU_Sales','JP_Sales','Other_Sales','Critic_Count','User_Count', 'Rating'], axis=1)
```

```python
display(df)
```

Verificando valores únicos, dessa forma podemos diferenciar as quantidades de plataformas e generos como outras informações: 

```python
df.nunique()
```

Abaixos podemos ver que a principal plataforma é o Playstation 2 (PS2 seguido do Nintendo DS (DS)).

```python
df['Platform'].value_counts()
```

Verificando as principais Publisher que são as empresas que publicam os jogos mas não necessáriamente o desenvolvem.

```python
df['Publisher'].value_counts()
```

Verificando as principais Desenvolvedoras.

```python
df['Developer'].value_counts()
```

Abaixo vou remover as linhas onde nas colunas (Name, Platform, Year_of_Release, Genre, Global_Sales) tenha algum dado ausente: 

```python
df = df.dropna(subset=['Name', 'Platform', 'Year_of_Release', 'Genre', 'Global_Sales'])
```

```python
display(df)
```

Como ainda existem muitas linhas no arquivos e dados para serem trabalhados vamos fazer a limpeza dos dados não encontrados nas colunas Critic_Score User_Score

```python
df = df.dropna(subset=['Critic_Score', 'User_Score'])
```

```python
display(df)
```
